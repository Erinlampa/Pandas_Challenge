```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#make count for passing scores in new column
student_data['pass_reading'] = np.where(student_data['reading_score']>=70, 1, 0)
student_data['pass_math'] = np.where(student_data['math_score']>=70, 1, 0)
student_data.columns

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools at district level
total_schools = school_data["school_name"].count()

#Calculate the total number of students
total_students = school_data_complete["Student ID"].count()

#Calculate the total budget
total_budget = school_data["budget"].sum()

#Calculate the average math score
avg_math = int(school_data_complete["math_score"].mean())

#Calculate the average reading score
avg_reading = int(school_data_complete["reading_score"].mean())

#Calculate the percentage of students with a passing math score (70 or greater)
pass_math = (school_data_complete["math_score"] > 69) *100
pm = (pass_math.sum()/total_students)

#Calculate the percentage of students with a passing reading score (70 or greater)
pass_reading = (school_data_complete["reading_score"] > 69) *100
pr = (pass_reading.sum()/total_students)
avg = (pm + pr)/2
```

```python
#Optional: give the displayed data cleaner formatting
d = {'Total Schools' : [total_schools], "Total Students" : [total_students],
     "Total Budget" :[total_budget],"Avg Math Score" : [avg_math],"Avg Reading Score" : [avg_reading],
     "% Passing Math": [pm], "% Passing Reading" :[pr], "Overall % Pass" :[avg]} 
district_summary = pd.DataFrame(data=d)
district_summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#Create a group Sum
group_sum = student_data.groupby('school_name', as_index=False).sum()

#Groupby school
group_mean = student_data.groupby('school_name', as_index=False).mean()
group_mean = group_mean[['school_name', 'reading_score', 'math_score']]
```

```python
#Merge the data
merged_data = pd.merge(group_sum,group_mean, on='school_name')
school_summary = pd.merge(school_data, merged_data, left_on='school_name', right_on='school_name')

#Calculate budget per student
per_student = school_summary['budget']/school_summary['size']
school_summary['Budget Per Student'] = per_student

#Calculate pass reading and math
passreading = school_summary['pass_reading'] = school_summary['pass_reading']/school_summary['size']*100
passmath = school_summary['pass_math'] = school_summary['pass_math']/school_summary['size']*100

#Calculate overall passing rate
school_summary['Overall Passing Rate'] = (passreading + passmath)/2
```

```python
#Rename Columns to clean up
school_summary.rename(columns = {'reading_score_y':'Average Reading Score',
                                 'math_score_y':'Average Math Score', 
                                 'pass_reading':'% Passing Reading',
                                 'pass_math':'% Passing Math', 'school_name':'School',
                                 'type':'Type', 'Size':'Enrollment',
                                 'budget':'Budget'}, inplace = True)
school_summary = school_summary[['School', 'Type', 'Budget', 'size', 'Budget Per Student',
       '% Passing Reading','% Passing Math', 'Average Reading Score', 'Average Math Score',
       'Overall Passing Rate']]
school_summary
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe

```python
#Group by schools and Grade
# Apply mean function for reading & math scores
students_mean_df = student_data.groupby(['school_name','grade'], as_index=False).mean()
students_mean_df.sort_values(by=['school_name','grade'], ascending=[1,1]).head()
students_mean_df= students_mean_df[['school_name', 'grade', 'Student ID', 'reading_score', 'math_score',
                                    'pass_reading', 'pass_math']]
students_mean_df

#Create pivot table and re-order grades

table_math = pd.pivot_table(students_mean_df, values='pass_math', index=['school_name'],
                 columns=['grade'], aggfunc=np.mean)

#Change values to percent
table_math['10th'] = table_math['10th']*100
table_math['11th'] = table_math['11th']*100
table_math['12th'] = table_math['12th']*100
table_math['9th']  = table_math['9th']*100

#rearrnge grades to crrect order

table_math = table_math[['9th', '10th', '11th', '12th']]
table_math
```

```python
#Scores by School Spending
bins = [0, 585, 615, 645, 675]
labels = ["<$585", "$585-615", "$615-645", "$645-675"]
```
